```python
#can get info for the function by adding the ? at the end
stats.ttest_ind?
```

```python
# can do other tests as well, this one with a made up 2x2 contingency table
fisher_result = stats.fisher_exact([[20,5],
                                    [8,30]])
```

# Statsmodels

```python
# using statsmodels (sm) to get some functions we want
# we are loading in the california housing datafile. This is a tsv, so we need to specify it's tab separated

df = pd.read_csv("california_housing.tsv", sep="\t", index_col=0)
```

```python
df
```

```python
#OLS = ordinary least squares model. Provide first the response variable (median house value), then select specific with df.loc, so all rows and median income value
# should provide an intercept value, which there is a specific way to do so.
# add constant is a way for stats models to have an intercept term in your linear model (a column of 1 is equivelant to an intercept value) 
# we add sm.add_constant
california_ols = sm.OLS(df["MedHouseVal"], sm.add_constant(df.loc[:,["MedInc", "Latitude"]]))
```

```python
#the sm.add_constant adds this variable below to the dataframe
sm.add_constant(df.loc[:, ["MedInc"]])
```

```python
#fit 
california_ols_result = california_ols.fit()
```

```python
#print the results summary ran above
print(california_ols_result.summary())
```
